```python
import pandas as pd
import seaborn as sns
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('DataCities.csv',sep=',',index_col=0)
```

```python

df.head(3)
```

```python
df2d=df[['Casablanca','Marrakech','Country']]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df2d.drop(columns=['Country']))
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
scale=1

plt.rcParams['figure.figsize']=(8,8) #rescale figures
plt.rcParams.update({'font.size': 18})
sns.scatterplot(data=df2d,x='Casablanca',y='Marrakech',hue='Country')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Distance to Casablanca')
plt.ylabel('Distance to Marrakech')

plt.arrow(x=7500, y=7500, dx=pca.components_[0][0]*7000*scale,
          dy=pca.components_[0][1]*7000*scale,color='r',head_width=200,
         label='PC1')

plt.arrow(x=5000, y=5000, dx=pca.components_[1][0]*200*scale,
          dy=pca.components_[1][1]*200*scale,color='b',head_width=200,
         label='PC2')


plt.text(x=5000,y=5300,s='PC1',rotation=45,color='r')
plt.text(x=5200,y=4750,s='PC2',rotation=-45,color='b')
```

```python
sns.scatterplot(x=pd.Series(pca.explained_variance_ratio_,index=['PC1','PC2']).index,
            y=pd.Series(pca.explained_variance_ratio_,index=['PC1','PC2']).values,
               s=500)
plt.ylabel('Fraction of variance explained')

```
